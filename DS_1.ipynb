```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quanQual(dataset)
```

```python
quan,qual=quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
dataset["salary"].fillna(0,inplace=True)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df = pd.DataFrame(df,columns = quan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy="most_frequent")
imp.fit(dataset[qual])
df1=imp.transform(dataset[qual])
df1 = pd.DataFrame(df1,columns =qual)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
dataset1=[df,df1]
```

```python
dataset1=pd.concat(dataset1,axis=1)
```

```python
dataset1
```

```python
dataset1.isnull().sum()
```

```python
dataset1.to_csv("placement_data_.csv",index=False)
```
